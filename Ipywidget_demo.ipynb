```python
import ipywidgets as widgets
from IPython.display import display
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib nbagg
```

$\psi = A \sin(2 \pi \nu x)$

where
$A = Amplitude$, $\nu = Frequency$


```python
x = np.linspace(0,10,1001)
fig, ax = plt.subplots(1, figsize=(10,6))
plt.suptitle('Interactive demo')

def update_plot(amp, freq):
    '''
    This function is linked to the sliders and 
    replots the function as sliders are modified.
    '''
    ax.clear(); 
    y = amp*np.sin(2*np.pi*freq*x)
    ax.plot(x,y)

a = widgets.FloatSlider(min = 1.0, max = 10.0, value = 1.0, description = 'Amplitude')
f = widgets.FloatSlider(min = 1.0, max = 10.0, value = 1.0, description = 'Frequency')

widgets.interactive(update_plot, amp = a, freq = f )
```
